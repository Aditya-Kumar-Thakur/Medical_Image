```python
import pandas as pd

# Provide the path to your CSV file
file_path = r"C:\Users\adity\combined_rows.csv"

# Read the CSV file
data = pd.read_csv(file_path)

# Display the first few rows of the data
print(data.head())
```

```python
data
from sklearn.preprocessing import LabelEncoder
Output=LabelEncoder()
data['Output']=Output.fit_transform(data['Output'])
```

```python
features_cols=['White Matter', 'Grey Matter',
                 'CSF', 'Accumbens', 'Amygdala', 'Caudate', 'Hippocampus',
                 'Putamen']
X=data[features_cols]
Y=data.Output
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier=DecisionTreeClassifier(criterion="entropy")
```

```python
classifier.fit(X_train,Y_train)
```

```python
Y_pred=classifier.predict(X_test)
Y_pred
```

```python
X_test
```

```python
classifier.score(X_test,Y_test)
```

```python
# Evaluate the model
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
#from sklearn.metrics import accuracy_score
train_accuracy = accuracy_score(Y_train, classifier.predict(X_train))
test_accuracy = accuracy_score(Y_test, classifier.predict(X_test))

print(f"Training Accuracy: {train_accuracy:.2f}")
print(f"Testing Accuracy: {test_accuracy:.2f}")
```

```python
import pandas as pd
from sklearn.feature_selection import mutual_info_classif
import numpy as np

# Load the dataset
file_path = r"C:\Users\adity\combined_rows.csv"  # Update with your file path
data = pd.read_csv(file_path)

# Drop unnecessary columns (e.g., 'Subject', columns with null values)
data_cleaned = data.drop(columns=['Subject', 'Basal Forebrain', 'Pallidum']).dropna()

# Separate features (X) and target (y)
X = data_cleaned.drop(columns=['Output']).values
y = data_cleaned['Output']

# Calculate mutual information (approximation of information gain)
mutual_info = mutual_info_classif(X, y, discrete_features=False)

# Map mutual information values to their corresponding feature names
features = data_cleaned.drop(columns=['Output']).columns
feature_importance = dict(zip(features, mutual_info))

# Sort features by their importance
sorted_features = sorted(feature_importance.items(), key=lambda item: item[1], reverse=True)

# Print sorted features and their information gain
print("Feature Importance based on Information Gain:")
for feature, importance in sorted_features:
    print(f"{feature}: {importance:.4f}")
```
